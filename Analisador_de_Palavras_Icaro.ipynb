# IMPORTA√á√ÉO DAS BIBLIOTECAS NECESS√ÅRIAS


```python
import string                       # Para manipula√ß√£o de pontua√ß√£o
import collections                  # Para contar frequ√™ncia das palavras
from wordcloud import WordCloud     # Para gerar a nuvem de palavras
import matplotlib.pyplot as plt     # Para plotar gr√°ficos
from nltk.corpus import stopwords   # Lista de palavras comuns a serem removidas
import nltk                         # Toolkit para processamento de linguagem natural
from google.colab import files      # Para upload de arquivos no Colab
import io                           # Para manipular arquivos
```

# DOWNLOAD DAS STOPWORDS (Necess√°rio na primeira execu√ß√£o)

```python
nltk.download('stopwords')
```

# ENTRADA DE DADOS

```python
# O usu√°rio pode digitar manualmente ou fazer upload de um arquivo .txt
modo = input("Digite 1 para inserir frases manualmente ou 2 para enviar um arquivo .txt: ")

if modo == "1":
    # Entrada manual
    texto = input("Digite suas frases: ")
elif modo == "2":
    # Upload do arquivo
    uploaded = files.upload()  # Abre o seletor de arquivos no Colab
    for file in uploaded.values():
        texto = file.decode("utf-8")  # Converte o conte√∫do para string
else:
    print("Op√ß√£o inv√°lida. Reinicie e digite 1 ou 2.")
    texto = ""
```

# PR√â-PROCESSAMENTO DO TEXTO

```python
# Cria√ß√£o de um "tradutor" para remover pontua√ß√µes
tradutor = str.maketrans('', '', string.punctuation)

# Remove pontua√ß√£o e coloca tudo em min√∫sculas
texto_limpo = texto.translate(tradutor).lower()

# Separa o texto em palavras
palavras = texto_limpo.split()
```

# REMO√á√ÉO DE STOPWORDS

```python
# Carrega a lista de stopwords em portugu√™s
stop_words = set(stopwords.words('portuguese'))

# Remove palavras comuns (ex: "de", "a", "e", etc.)
palavras_filtradas = [palavra for palavra in palavras if palavra not in stop_words]
```

# CONTAGEM DE FREQU√äNCIA DAS PALAVRAS

```python
frequencia = collections.Counter(palavras_filtradas)

# Pergunta ao usu√°rio quantas palavras mais frequentes ele deseja visualizar
n = int(input("Quantas palavras mais frequentes voc√™ quer ver? "))

# Obt√©m as N palavras mais comuns
mais_comuns = frequencia.most_common(n)

# Exibe as palavras e suas quantidades
print("\nPalavras mais frequentes:")
for palavra, contagem in mais_comuns:
    print(f"{palavra}: {contagem}")
```

# GERA√á√ÉO DA NUVEM DE PALAVRAS

```python
# Cria a nuvem com base nas frequ√™ncias
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(frequencia)

# Exibe a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Nuvem de Palavras")
plt.show()
```

# GR√ÅFICO DE BARRAS DAS PALAVRAS MAIS FREQUENTE

```python
# Extrai os dados para o gr√°fico
palavras_top = [item[0] for item in mais_comuns]
contagens_top = [item[1] for item in mais_comuns]

# Plota o gr√°fico de barras
plt.figure(figsize=(10, 5))
plt.bar(palavras_top, contagens_top, color='skyblue')
plt.title("Frequ√™ncia das Palavras Mais Comuns")
plt.xlabel("Palavras")
plt.ylabel("Frequ√™ncia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
